```python
### create a function of getting the position of the reference point we set 
def quick_sort_part(array, start, end): 
  orig_pivot = array[start]
  right_pointer = end
  left_pointer = start+1

  if len(array) == 1:
    return
  while True:
    # if the right pointer > left pointer and the element in the right pointer > reference point, then the right pointer could decrease
    while left_pointer <= right_pointer and array[right_pointer] > orig_pivot:
      right_pointer -= 1
    # if the right pointer > left pointer and the element in the left pointer < reference point, then the left pointer could increase  
    while left_pointer <= right_pointer and array[left_pointer] < orig_pivot:
      left_pointer += 1
    # based on the condition : right pointer >= left pointer, the elements in those two position should switch 
    if left_pointer <= right_pointer: 
      array[left_pointer], array[right_pointer] = array[right_pointer], array[left_pointer]
    else:
      break
  # based on the final right pointer, we can find the new position of the reference point    
  array[start], array[right_pointer] = array[right_pointer], array[start]
  return right_pointer

### create a function of splitting the array based on the new position of reference point 
def quick_sort(array, start, end):
  if start >= end:
    return  
  part_index = quick_sort_part(array, start, end) # get the position of reference point
  quick_sort(array, start, part_index-1) # left array of the previous position of reference point and keep recycling the finding of reference point until the start point < end point 
  quick_sort(array, part_index+1, end) # right array of the previous position of reference point and keep recycling the finding of reference point until the start point < end point 

if __name__ == '__main__':
  import numpy as np
  Ori_array = np.array([ 33, 67, 8, 13, 54, 119, 3, 84, 25, 41])
  quick_sort(Ori_array, start = 0, end = len(Ori_array)-1)
  print("The new sort of the array inputted = ", Ori_array)
```
